```python
import pandas as pd
from scipy.stats import chi2_contingency

# Asegúrate de que ya tienes tu DataFrame original 'df'
# Y que la columna 'facultad' está en tu DataFrame.

# Seleccionar las columnas de interés (p1 a p18)
columnas_preguntas = [f'p{i}' for i in range(1, 19)]

print("Realizando pruebas de Chi-Cuadrado de Independencia:")
print("-" * 50)

# Diccionario para almacenar los resultados de las pruebas
resultados_chi2 = {}

# Realizar la prueba de chi-cuadrado para cada columna de pregunta
for columna in columnas_preguntas:
    if columna in df.columns and 'facultad' in df.columns:
        print(f'Analizando columna: {columna}')

        # Crear una tabla de contingencia
        # Esto cuenta la frecuencia de cada combinación de respuesta y facultad
        tabla_contingencia = pd.crosstab(df['facultad'], df[columna])

        # Asegurarse de que la tabla de contingencia no esté vacía
        if not tabla_contingencia.empty and (tabla_contingencia.sum().sum() > 0):
            try:
                # Realizar la prueba de chi-cuadrado
                # chi2: el estadístico de chi-cuadrado
                # p: el valor p
                # dof: grados de libertad
                # expected: las frecuencias esperadas bajo la hipótesis nula
                chi2, p, dof, expected = chi2_contingency(tabla_contingencia)

                print(f'  Estadístico Chi-cuadrado: {chi2:.4f}')
                print(f'  Valor p: {p:.4f}')

                # Interpretar el valor p (usando un nivel de significancia alfa = 0.05)
                alpha = 0.05
                if p < alpha:
                    print(f'  Resultado: Diferencia estadísticamente significativa entre las facultades (p < {alpha}).')
                    resultados_chi2[columna] = {'chi2': chi2, 'p_value': p, 'resultado': 'Significativo'}
                else:
                    print(f'  Resultado: No hay diferencia estadísticamente significativa entre las facultades (p >= {alpha}).')
                    resultados_chi2[columna] = {'chi2': chi2, 'p_value': p, 'resultado': 'No significativo'}

            except ValueError as e:
                 print(f"  Error al realizar la prueba de chi-cuadrado para {columna}: {e}")
                 print("  Esto puede ocurrir si hay filas/columnas con todos los valores cero en la tabla de contingencia.")
                 resultados_chi2[columna] = {'chi2': None, 'p_value': None, 'resultado': 'Error o datos insuficientes'}

        else:
            print(f'  Tabla de contingencia vacía o con suma cero para la columna {columna}. No se puede realizar la prueba.')
            resultados_chi2[columna] = {'chi2': None, 'p_value': None, 'resultado': 'Tabla vacía'}

        print("-" * 30)
    elif columna not in df.columns:
        print(f'La columna {columna} no se encuentra en el DataFrame.')
        print("-" * 30)
    else:
        print(f'La columna "facultad" no se encuentra en el DataFrame.')
        print("-" * 30)

print("\nResumen de resultados de la prueba de Chi-Cuadrado:")
for columna, resultado in resultados_chi2.items():
    print(f"  {columna}: Valor p = {resultado['p_value']:.4f}, Resultado = {resultado['resultado']}")
```
